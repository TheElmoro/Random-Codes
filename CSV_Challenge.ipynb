```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('/content/drive/MyDrive/Book1.csv')

print(dataset.head())
```

```python
plt.figure()
plt.hist(dataset["age"], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
counts = dataset["nationality"].value_counts().sort_values()

plt.figure()
plt.barh(counts.index, counts.values)
plt.title("Nationality Counts")
plt.xlabel("Count")
plt.ylabel("Nationality")
plt.show()
```

```python

gender_num = []
for g in dataset["gender"]:
    if str(g).strip().lower() == "male":
        gender_num.append(0)
    else:
        gender_num.append(1)

plt.figure()
plt.scatter(gender_num, dataset["age"])
plt.title("Age vs Gender")
plt.xlabel("Gender (0 = Male, 1 = Female)")
plt.ylabel("Age")
plt.show()
```
