```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv("/content/gdrive/MyDrive/Book1.csv")
```

```python
dataset.head()
```

```python
print(dataset.isna().sum())
print()
print(dataset.describe)
print()
print(dataset.info())
```

```python
plt.figure()
plt.scatter(dataset["gender"], dataset["nationality"])
plt.title("Nationality and Gender")
plt.xlabel("Gender")
plt.ylabel("Nationality")
plt.show()
```

```python
counts = dataset["nationality"].value_counts().sort_values()
plt.figure()
plt.barh(counts.index, counts.values)
plt.title("Nationality Counts")
plt.xlabel("Count")
plt.ylabel("Nationality")
plt.savefig("/content/gdrive/MyDrive/nationality_gender.png", dpi=150, bbox_inches="tight")
plt.show()
```
